```python
TotalLayoffs = df_copy['total_laid_off'].sum()

Layoffsbycountry = df_copy.groupby('country')['total_laid_off'].sum().reset_index()



Layoffsbycountry['percentagelaid_off'] = (Layoffsbycountry['total_laid_off'] / TotalLayoffs ) * 100


Layoffsbycountry['percentagelaid_off'] = Layoffsbycountry['percentagelaid_off'].round(0)

Layoffsbycountry = Layoffsbycountry.sort_values(by='percentagelaid_off', ascending=False)

Layoffsbycountry
```
